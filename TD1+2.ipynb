## 1. Contexte général

La conduite automobile est une activité complexe qui mobilise plusieurs ressources cognitives : perception visuelle, attention soutenue, prise de décision, anticipation, contrôle moteur et gestion des imprévus.

Le papier CL-Drive s'intéresse donc à une question centrale :

> Peut-on estimer automatiquement la charge cognitive d'un conducteur à partir de signaux physiologiques et comportementaux mesurés pendant la conduite ?

#### Q1. Pourquoi est-il intéressant d'estimer la charge cognitive d'un conducteur ?

**Réponse :**
If the cognitive load increases beyond a certain point, the individual's performance will degrade, which in case of driving may increase the likelihood of road accidents.


## 2. Définition : charge cognitive

#### Q2. Comment définir la charge cognitive ?

**Réponse :**
Cognitive load refers to the quantity of information our working memory can process at a given time. In other words, it is the amount of cognitive resources required to accomplish a task.


#### Q3. Dans le contexte de la conduite, de quels facteurs la charge cognitive peut-elle dépendre ?

**Réponse :**

In general, two categories of cognitive load, intrinsic and extraneous, have been described in the literature. While intrinsic cognitive load is defined as the inherent complexity of a given task, extraneous cognitive load refers to the cognitive resources demanded by environmental factors that are task-irrelevant.

Dans le papier, la charge cognitive est évaluée à l'aide d'une échelle subjective appelée **PAAS**, allant de 1 à 9.

## 5. Modalités mesurées

Le dataset CL-Drive est **multimodal**. Il contient plusieurs types de signaux enregistrés pendant la même tâche de conduite.

#### Q7. Compléter le tableau suivant.

| Modalité | Signification | Signal mesuré | Rôle possible |
|---|---|---|---|
| EEG | Electroencephalogram | Activité électrique cérébrale (4 canaux : AF7, AF8, TP9, TP10) à 256 Hz |Modalité principale BCI |
| ECG | Electrocardiography | Activité électrique cardiaque (LL-RA, LA-RA, Vx-RA) à 512 Hz |Variabilité cardiaque  |
| EDA | Electrodermal Activity | Conductance cutanée, 2 électrodes abdomen, 128 Hz |Réponse du système nerveux sympathique |
| Gaze | Eye tracking | Saccades, fixations, diamètre pupillaire, clignements, à 50 Hz |Comportement oculaire lié à la charge|

Dans le projet, la modalité principale sera l’**EEG**. Les autres modalités serviront éventuellement d’extensions pour discuter de la fusion multimodale.

#### Q8. Pourquoi peut-il être intéressant de combiner EEG, ECG, EDA et Gaze pour estimer la charge cognitive ?

**Réponse :**
Multimodal approaches have proven effective at magnifying the accuracy of cognitive load assessment in the past. Apart from EEG, research has shown that both the sympathetic nervous system (SNS) which controls the skin conductance response and automatic nervous system which controls the heart rate variability (HRV) are impacted by cognitive load. Prior research has also shown significant correlation between changes in pupil size, blink rate, saccade, and fixation with cognitive load.

## 7. Protocole expérimental

#### Q11. Quelle est la différence entre un **protocole expérimental** et un **dataset exploitable en machine learning** ?

**Réponse :**

le protocole expérimental décrit les conditions d'acquisition (capteurs, scénarios, sujets, timing), tandis que le dataset ML désigne les données effectivement exploitables après nettoyage.

#### Q12. Donner une description détaillée du protocole expérimental utilisé.

**Réponse :**

First, 3 minutes of baseline data was collected from each participant. participants were asked to perform the Orientation scenario. Following every 3-minute driving scenario, the participants were given a resting time of 2 minutes to allow them to rest, reduce the possibility of SAS, and come back to a relatively lower cognitive load state. Followed by the resting period, a 2-minute baseline was collected before each new scenario.

#### Q13. Pourquoi les auteurs collectent-ils une baseline avant les scénarios ?

**Réponse :**

To reduce the variability between subjects, which is a common phenomenon when recording such data, we divide each feature value with its corresponding average value from the baseline.

#### Q14. Pourquoi prévoir un scénario d’orientation avant les scénarios principaux ?

**Réponse :**
To become adapted to the simulated driving environment and make sure that participants had a clear understanding about the 'low', 'medium', and 'high' complexity levels, and also to reduce the SAS level, participants were asked to perform the Orientation scenario.


#### Q15. Pourquoi prévoir des périodes de repos entre les scénarios ?

### Réponse

the participants were given a resting time of 2 minutes to allow them to rest, reduce the possibility of SAS, and come back to a relatively lower cognitive load state.

#### Q16. Pourquoi demander un score de charge cognitive toutes les 10 secondes plutôt qu’une seule fois à la fin du scénario ?

### Réponse

Through a few pilot trials, 10 seconds was found to be a reasonable interval, as longer intervals could lead participants to forget their earlier experiences during that segment, while shorter intervals could interfere with the experiment itself.

### Réponse
We observe that as the scenarios progress, more and more participants select higher cognitive load scores, indicating that the task complexities do indeed increase as the scenarios progress, especially for scenarios 8 and 9.


#### Q20. La complexité du scénario suffit-elle à définir le label de charge cognitive ?

### Réponse
Participant cognitive load self-assessment ratings were used as ground truth labels in this study.


## 9. Échantillonnage et segmentation temporelle

Les signaux EEG sont continus. Pour entraîner un modèle, il faut transformer ce signal continu en exemples individuels.

Dans CL-Drive, les scores PAAS de charge cognitive sont collectés toutes les 10 secondes.

Il est donc naturel de découper les signaux physiologiques en fenêtres de 10 secondes :

$$
X_i = \text{segment de 10 s}
$$

$$
y_i = \text{score PAAS associé}
$$

On obtient ainsi le problème suivant :

Le principe d’association segment-label est le suivant :

| Temps | Segment EEG | Score PAAS |
|---|---|---|
| 0–10 s | segment 1 | score 1 |
| 10–20 s | segment 2 | score 2 |
| 20–30 s | segment 3 | score 3 |

#### Q21. Quelle est la fréquence d’échantillonnage EEG utilisée dans le papier ?


**Réponse :**

The sampling rate of the EEG headband is 256 Hz.

#### Q22. Combien d’échantillons contient une fenêtre EEG de 10 secondes ?


**Réponse :**
256 Hz × 10 s = 2560 échantillons

#### Q23. Combien de valeurs contient une fenêtre EEG de 10 secondes si les 4 canaux sont utilisés ?

**Réponse :**

2560 × 4 = 10 240 valeurs.

#### Q24. Un scénario dure 3 minutes. Combien de fenêtres de 10 secondes peut-on obtenir par scénario ?

### Réponse

3 minutes = 180 s → 180 / 10 = 18 fenêtres.

#### Q25. Si un sujet réalise 9 scénarios complets, combien de fenêtres obtient-on pour ce sujet ?

### Réponse

18 × 9 = 162 fenêtres par sujet.

#### Q26. Si 21 sujets réalisent tous 9 scénarios complets, combien de fenêtres obtient-on au total ?

### Réponse

162 × 21 = 3 402 fenêtres

## 10. Formulation du problème d'apprentissage automatique

À partir du dataset CL-Drive, on peut formuler plusieurs problèmes de classification.

### Classification binaire

| Scores PAAS | Classe |
|---|---|
| 1 à 4 | charge faible |
| 5 à 9 | charge élevée |

### Classification ternaire

| Scores PAAS | Classe |
|---|---|
| 1 à 3 | charge faible |
| 4 à 6 | charge moyenne |
| 7 à 9 | charge élevée |

Dans le projet, on commencera par la classification binaire.

#### Q27. Pourquoi commencer par une classification binaire plutôt qu’une classification ternaire ?

### Réponse
ternary classification is more challenging than binary


#### Q28. Dans une classification binaire, pourquoi le score PAAS 5 est-il généralement placé dans la classe « charge élevée » dans ce papier ?

### Réponse

For binary, we group the cognitive load ratings from 1 to 4 as 'low' cognitive load and 5 to 9 as 'high' cognitive load.

#### Q29. Quel est le lien entre ce papier et un système d'interface adaptative ?

**Réponse :**

recent intelligent technologies integrated into vehicles should possess the ability to measure cognitive load and alarm the user should dangerously high amounts of it be detected.

## 9. Visualisation des signaux EEG

Avant tout prétraitement ou apprentissage automatique, il est indispensable de visualiser les signaux EEG.

La visualisation permet de :

- vérifier que les données sont correctement chargées ;
- observer les amplitudes ;
- repérer les valeurs aberrantes ;
- détecter les ruptures ou données manquantes ;
- comparer les canaux ;
- repérer des artefacts évidents ;
- comparer qualitativement des segments de faible et forte charge cognitive.

### Algorithme conceptuel pour visualiser un segment EEG

Pour visualiser un segment EEG de 10 secondes :

1. choisir un sujet ;
2. choisir un scénario ;
3. choisir un segment ;
4. récupérer les 4 canaux EEG ;
5. calculer le nombre d’échantillons : $N = f_s \times T$ ;
6. construire l’axe temporel : $t[n] = n/f_s$ ;
7. tracer les quatre canaux en fonction du temps ;
8. ajouter le score PAAS associé au titre du graphique ;
9. commenter la qualité du signal.

#### Q30. Que faut-il regarder en premier lorsqu’on affiche un signal EEG brut ?

### Réponse

amplitude, présence d'artefacts, données manquantes, cohérence entre canaux.

#### Q31. Pourquoi afficher les quatre canaux EEG plutôt qu’un seul ?

### Réponse

les régions frontales (AF7/AF8) et temporales (TP9/TP10) capturent des informations différentes

#### Q32. Pourquoi comparer visuellement un segment de faible charge et un segment de forte charge cognitive ?

### Réponse

#### Q33. Pourquoi l'utilisation d'un simulateur est-elle à la fois un avantage et une limite ?

**Réponse :**
using a driving simulator versus real vehicles offers a safe alternative, reduces the risk of accidents and injuries, and allows us to design very specific driving scenarios by controlling the weather, time of day, road conditions. On the other hand, the disadvantage of using a simulator is the possibility of participants experiencing SAS, as well as limited motion and the use of generated graphics that may not fully replicate the experience of real-world driving.

Ce papier présente une étude sur la mesure de la charge cognitive d'un conducteur pendant la conduite.
